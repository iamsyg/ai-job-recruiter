```python
importlib.reload(sys.modules['utils.skills'])
from utils.skills.skill_names import SKILL_TAXONOMY

counts = []
max_durations = []
avg_durations = []
endorsements = []

for feature in features:

    skill_groups = feature["skill_features"]

    total_count = sum(
        s["count"]
        for s in skill_groups.values()
    )

    max_duration = max(
        (s["max_duration"] for s in skill_groups.values()),
        default=0
    )

    durations = [
        s["avg_duration"]
        for s in skill_groups.values()
        if s["count"] > 0
    ]

    avg_duration = (
        sum(durations) / len(durations)
        if durations else 0
    )

    total_endorsements = sum(
        s["endorsements"]
        for s in skill_groups.values()
    )

    counts.append(total_count)
    max_durations.append(max_duration)
    avg_durations.append(avg_duration)
    endorsements.append(total_endorsements)


print("\nSkill Count")
print(np.percentile(counts, [10,25,50,65,70,75,90,95]))

print("\nSkill Max Duration")
print(np.percentile(max_durations, [10,25,50,65,70,75,90,95]))

print("\nSkill Avg Duration")
print(np.percentile(avg_durations, [10,25,50,65,70,75,90,95]))

print("\nSkill Endorsements")
print(np.percentile(endorsements, [10,25,50,65,70,75,90,95]))


for group in SKILL_TAXONOMY:

    counts = [
        f["skill_features"][group]["count"]
        for f in features
    ]

    print(group)
    print(np.percentile(counts, [50,75,90,95]))
```

```python
import numpy as np

from utils.load_candidates import load_candidates




def analyze_job_duration(candidates):

    durations = []

    for candidate in candidates:
        for role in candidate.career_history:
            durations.append(role.duration_months)

    durations = np.array(durations)

    print("Min:", durations.min())
    print("Mean:", durations.mean())
    print("Median:", np.median(durations))

    print("25% :", np.percentile(durations, 25))
    print("50% :", np.percentile(durations, 50))
    print("75% :", np.percentile(durations, 75))
    print("90% :", np.percentile(durations, 90))
    print("95% :", np.percentile(durations, 95))

candidates = load_candidates(r"C:\Users\Eternity\Dropbox\Projects\New folder\[PUB] India_runs_data_and_ai_challenge\[PUB] India_runs_data_and_ai_challenge\India_runs_data_and_ai_challenge\candidates.jsonl")

analyze_job_duration(candidates)
```
